```python
import numpy as np
from PIL import Image
# !pip uninstall 'gast==0.2.2'
# !pip install 'gast==0.3.2'
# !pip install 'tensorflow==2.0.0'
# !pip install ISR
# !pip install h5py==2.10.0
from ISR.models import RDN
from ISR.models import RRDN
import os
import os.path
import shutil

ALLOWED_EXTENSIONS = {'png', 'jpg', 'jpeg'}

# os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

# This is a sample Python script.

#   https://www.mdeditor.tw/pl/2P3x/zh-hk
#   https://github.com/idealo/image-super-resolution/issues/85
#   https://github.com/Pulkitdzrt/Resolutionize
# Press Shift+F10 to execute it or replace it with your code.
# Press Double Shift to search everywhere for classes, files, tool windows, actions, and settings.

def print_hi(name):
    # Use a breakpoint in the code line below to debug your script.
    print(f'Hi, {name}')  # Press Ctrl+F8 to toggle the breakpoint.

def criaDir (caminho):
    retorno = False
    try:
        os.makedirs(caminho, mode=0o777, exist_ok=False)
    except OSError:
        print('Creation of the directory "%s" failed' % caminho)
    else:
        print('Successfully created the directory "%s"' % caminho)
        retorno = True
    return retorno

def handler(exc_info):
    print("Inside handler")
    print(exc_info)

def remocao (caminho):
    try:
        #   shutil.rmtree(caminho, ignore_errors=True)
        shutil.rmtree(caminho, onerror=handler)
    except OSError as e:
        print("Error: %s : %s" % (caminho, e.strerror))

def verificaDir(diretorio, erase):
    var = criaDir(diretorio)
    if (not var and erase):
         remocao(diretorio)
         var = criaDir(diretorio)
    return var

def novoDirRenomeia (caminho, diretorio, erase, div, width, height):
    print("\nRenomeando...")

    var = verificaDir(diretorio, erase)

    if (var):
        for pasta in os.listdir(caminho):
            subpasta = os.path.join(caminho, pasta)
            cSP = format(os.path.basename(subpasta))
            dest_dir = diretorio+"/"+cSP  #   cria a sub pasta no destino
            criaDir(dest_dir)
            #   get the current working dir
            #   src_dir = os.getcwd()
            for image_file in os.listdir(subpasta):
                fsp = image_file.split(".")
                ext = fsp[len(fsp) - 1]
                #   ext = "."+fsp[len(fsp)-1]
                if ext in ALLOWED_EXTENSIONS:
                    src_file = os.path.join(subpasta, image_file)
                    shutil.copy(src_file, dest_dir)  # copy the file to destination dir

                    dst_file = os.path.join(dest_dir, image_file)
                    os.chmod(dst_file, 0o777)

                    img = Image.open(dst_file)
                    iSR(dst_file, img, div, width, height);
                    print("\n...\n")


def iSR (arq, img, div, width, height):
  print("ISR on image:\t"+arq+"\t- Beginning")
  lr_img = np.array(img)
  
  model = RRDN(weights='gans')
  
  imgTemp = lr_img

  width0, height0 = Image.fromarray(imgTemp).size

  
  while width0<width and height0<height:
    print("RDN: gans")
    sr_img = model.predict(imgTemp)
    imgTemp = Image.fromarray(sr_img)
    width0, height0 = imgTemp.size
    imgTemp = np.asarray(imgTemp)

  imgTemp = Image.fromarray(imgTemp)
  width0, height0 = imgTemp.size

  if (width < width0 and height < height0):
    imgTemp = imgTemp.resize((width, height))

  print("RDN: psnr-small")
  rdn = RDN(weights='psnr-small')
  sr_img = rdn.predict(np.asarray(imgTemp))
  imgTemp = Image.fromarray(sr_img)
  width0, height0 = imgTemp.size

  if (width < width0 and height < height0):
    imgTemp = imgTemp.resize((width, height))

  print("RDN: noise-cancel")
  noise = RDN(weights='noise-cancel')
  sr_img = noise.predict(np.asarray(imgTemp))
  imgTemp = Image.fromarray(sr_img)
  width0, height0 = imgTemp.size
  
  if (width < width0 and height < height0):
    imgTemp = imgTemp.resize((width, height))

  print("ISR done")

  imgTemp.save(arq, "JPEG", optimize=True)
  print("Arquivo:\t"+arq+" - Salvo")
                    

# Press the green button in the gutter to run the script.
if __name__ == '__main__':
    
    dir = '/content/drive/MyDrive/IC/colab/s106x80'
    novoDirRenomeia (dir, 'output', True, 6, 640, 480)
    print("FIM")


# See PyCharm help at https://www.jetbrains.com/help/pycharm/
```
